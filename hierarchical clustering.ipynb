```python
"""
there are two type of  hierarchical clustering:
1. Agglomerative: Bottom up approach
2. Divisive: Top down approach
""" 
```

```python
# we will focus on agglomerative 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
"""
Firstly, we will check if AgglomerativeClustering can cluster the sphere-like data set
"""
```

```python
data_num = 1000
```

```python
means = [[1, 2], [4,9], [9, 0]]
cov = [[1,0],[0,1]]
cluster_0 = np.random.multivariate_normal(means[0], cov, data_num)
cluster_1 = np.random.multivariate_normal(means[1], cov, data_num)
cluster_2 = np.random.multivariate_normal(means[2], cov, data_num)
all_data = np.concatenate((cluster_0, cluster_1, cluster_2))
labels = []
```

```python
clustering = AgglomerativeClustering(n_clusters=3).fit(all_data)
```

```python
labels = clustering.labels_
```

```python
plt.scatter(all_data[:,0],all_data[:,1],c=labels,cmap='rainbow')
```

```python
"""
As we can see AgglomerativeClustering can effectively cluster the globe-like data set
Next we will check if can handle non-globe dataset
"""
```

```python
x = np.linspace(0,12,data_num)
y = np.random.rand(data_num,)+4
cluster_0_1 = np.vstack((x,y)).T
all_data_01 = np.concatenate((cluster_0_1, cluster_1, cluster_2))
```

```python
clustering_01 = AgglomerativeClustering(n_clusters=3, linkage='average').fit(all_data_01)
labels_1= clustering_01.labels_
#plot result
plt.scatter(all_data_01[:,0],all_data_01[:,1],c=labels_1,cmap='rainbow')
```

```python
"""
As we can see the result is little better than k-means but not completely solve our problem"""
```
